```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("newGreedyComparisonCost0.7.txt", sep = ",")
df = pd.DataFrame(df)
df.columns = ["onlineBarista","physicalBarista", "processingTime", "arrivalRate", "alpha", "numCustomer", "allQueue", "greedy", "pessGreedy", "allLine"]


df['allLine'] = df['allLine'].str.strip(']').astype(float)
df['onlineBarista'] = df['onlineBarista'].str.strip('[').astype(float)


pd.to_numeric(df['onlineBarista'])
pd.to_numeric(df['physicalBarista'])


df['greedy/OPT'] = df['greedy'].astype(float)/df['allQueue'].astype(float)
df['pessGreedy/OPT'] = df['pessGreedy']/df['allQueue']
        


# for i in range(0,6):
#         df_plot = df[df['onlineBarista'] == i]
#         df_plot = df_plot[df_plot['physicalBarista'] == 5-i]
#         print("df_plot dataframe is below")
#         print(df_plot)
       
#         plt.plot(df_plot['numCustomer'], df_plot['allQueue'], label = "allQueue")
#         plt.plot(df_plot['numCustomer'], df_plot['greedy'], label = "greedy")
#         plt.plot(df_plot['numCustomer'], df_plot['pessGreedy'], label = "pessGreedy")
#         plt.plot(df_plot['numCustomer'], df_plot['allLine'], label = "allLine")
#         plt.legend(loc="upper left", fontsize = "5")

#         plt.title("Online barista: " + str(i) + ", Physical barista: " + str(5-i))
#         plt.show()

        
       

print(df.dtypes)

for i in range(0,6):
        df_plot = df[df['onlineBarista'] == i]
        df_plot = df_plot[df_plot['physicalBarista'] == 5-i]
        df_plot = df_plot[(df_plot['numCustomer'] >= 100) & (df_plot['numCustomer'] <= 200)]

        print(df['greedy/OPT'])

        plt.scatter(df_plot['numCustomer'], df_plot['greedy/OPT'], label = "greedy/OPT")
        # plt.scatter(df_plot['numCustomer'], df_plot['pessGreedy/OPT'], label = "pessGreedy/OPT")
        plt.legend(loc="upper left", fontsize = "5")

        ax = plt.gca()
        ax.set_ylim(bottom = 1)

        plt.title("Ratio between Greedy and OPT with " + str(i) + " online barista," + str(5-i) + " physical barista: ")
        plt.show()
```
