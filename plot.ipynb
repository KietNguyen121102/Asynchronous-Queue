```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("costData0.7.txt", sep = ",")
df = pd.DataFrame(df)
df.columns = ["onlineBarista","physicalBarista", "processingTime", "arrivalRate", "alpha", "numCustomer", "customerNo", "allQueue", "SPNE", "greedy", "pessGreedy", "allLine"]


df['allLine'] = df['allLine'].str.strip(']').astype(float)
df['onlineBarista'] = df['onlineBarista'].str.strip('[').astype(float)


pd.to_numeric(df['onlineBarista'])
pd.to_numeric(df['physicalBarista'])



df_plot = df[df['onlineBarista'] == 3]
df_plot = df_plot[df_plot['physicalBarista'] == 2]
print("df_plot dataframe is below")
print(df_plot)
       
plt.plot(df_plot['customerNo'], df_plot['allQueue'], label = "allQueue")
plt.plot(df_plot['customerNo'], df_plot['SPNE'], label = "SPNE")
plt.plot(df_plot['customerNo'], df_plot['greedy'], label = "greedy")
plt.plot(df_plot['customerNo'], df_plot['pessGreedy'], label = "pessGreedy")
plt.plot(df_plot['customerNo'], df_plot['allLine'], label = "allLine")
plt.legend(loc="upper left", fontsize = "5")

        #plt.title("Online barista: " + str(i) + ", Physical barista: " + str(5-i))
plt.show()

        
       

print(df.dtypes)

```
