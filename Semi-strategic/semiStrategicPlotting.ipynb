```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

df = pd.read_csv("semiStrategic0.7.txt", sep = ",")
df = pd.DataFrame(df)
df.columns = ["onlineBarista","physicalBarista", "processingTime", "arrivalRate", "alpha", "physicalOverOnline", "greedyPopulation" , "totalCost"]


df['totalCost'] = df['totalCost'].str.strip(']').astype(float)
df['onlineBarista'] = df['onlineBarista'].str.strip('[').astype(float)


pd.to_numeric(df['onlineBarista'])
pd.to_numeric(df['physicalBarista'])
pd.to_numeric(df['physicalOverOnline'])

# df['physicalOverOnline'].round(2)
uniques = df['physicalOverOnline'].unique()
print(uniques)

# for i in range(0,6):
for value in uniques:
        for i in range(0,6):
                df_plot = df[df['onlineBarista'] == i]
                df_plot = df_plot[df_plot['physicalBarista'] == 5-i]
        
                print(value)
                plot = df_plot[df_plot['physicalOverOnline'] == value]
                plt.plot(plot['greedyPopulation'], plot['totalCost'], label = 'Online Barista: ' + str(i))
                plt.legend(loc = "upper left", fontsize = "7")
                # plt.legend(plot['physicalOverOnline'])
                plt.title("Physical population: " + str(value))


        
        plt.show()
      
       
        #plt.title("Online barista: " + str(i) + ", Physical barista: " + str(5-i))



        
       

print(df.dtypes)
```
