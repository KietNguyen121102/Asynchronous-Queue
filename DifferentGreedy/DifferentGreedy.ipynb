```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("greedyComparison0.4.txt", sep = ",")
df = pd.DataFrame(df)
df.columns = ["onlineBarista","physicalBarista", "processingTime", "arrivalRate", "alpha", "customer", "omniscient", "optimistic", "pessimistic"]


df['pessimistic'] = df['pessimistic'].str.strip(']').astype(float)
df['onlineBarista'] = df['onlineBarista'].str.strip('[').astype(float)


pd.to_numeric(df['onlineBarista'])
pd.to_numeric(df['physicalBarista'])


for i in range(0, 6):
    df_plot = df[df['onlineBarista'] == i]
    df_plot = df_plot[df_plot['physicalBarista'] == 5-i]
    df_plot = df_plot[df_plot['customer'] <= 200]
    print("df_plot dataframe is below")
    print(df_plot)
       
    plt.plot(df_plot['customer'], df_plot['omniscient']/df_plot['customer'], label = "omniscient")
    plt.plot(df_plot['customer'], df_plot['optimistic']/df_plot['customer'], label = "optimistic")
    # plt.plot(df_plot['optimistic'], df_plot['customer'], label = "optimistic")
    plt.plot(df_plot['customer'], df_plot['pessimistic']/df_plot['customer'], label = "pessimistic")
   
    plt.title("Num of online baristas: " + str(i) + ", Num of physical baristas:" + str(5-i))
    plt.legend(loc="upper left", fontsize = "5")

        #plt.title("Online barista: " + str(i) + ", Physical barista: " + str(5-i))
    plt.show()

        
       

print(df.dtypes)

```
