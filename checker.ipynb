```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np 


df = pd.read_csv("greedyExpectedCostData.txt", sep = ",")
df.columns = np.arange(1,101)
x = np.arange(1,101)

f = plt.figure()
f.set_figwidth(15)

y = [float(df.iloc[0][i]) for i in range(2,100)]
y.insert(0,float(df.iloc[0][1][1:]))
y.insert(100,float(df.iloc[0][100][:-1]))
print(len(y))
print(y)

y1 = [float(df.iloc[1][i]) for i in range(2,100)]
y1.insert(0, float(df.iloc[1][1][1:]))
y1.insert(100, float(df.iloc[1][100][:-1]))

plt.scatter(x,y)
plt.scatter(x,y1)
plt.title("1 Online Barista, 1 Physical Barista, Processing Time: 3, Alpha: 0.9")

```
