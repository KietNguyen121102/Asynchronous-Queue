```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("newGreedyComparisonCost0.7.txt", sep = ",")
df = pd.DataFrame(df)
df.columns = ["onlineBarista","physicalBarista", "processingTime", "arrivalRate", "alpha", "numCustomer", "allQueue", "greedy", "pessGreedy", "allLine"]


df['allLine'] = df['allLine'].str.strip(']').astype(float)
df['onlineBarista'] = df['onlineBarista'].str.strip('[').astype(float)


pd.to_numeric(df['onlineBarista'])
pd.to_numeric(df['physicalBarista'])
df['greedy=allLine'] = df['numCustomer']


for i in range(0,6):
        df_plot = df[df['onlineBarista'] == i]
        df_plot = df_plot[df_plot['physicalBarista'] == 5-i]
        list = []
        for k in range(0,499):
            temp = df_plot['allLine'].iloc[k]
            for j in range(0,499):
                if (df_plot["greedy"].iloc[j] >= temp):
                    list.append(j)
                    break
        second_list = []
        for m in range(0, len(list)):
            second_list.append(m)
        plot_df = pd.DataFrame(list, columns = ['greedyEquivalence'])  
        plot_df['worstCase'] = second_list    
        plot_df['greedyEquivalence/worstCase'] = plot_df['greedyEquivalence']/plot_df['worstCase']
        print(plot_df) 
        plt.plot( plot_df['worstCase'], plot_df['greedyEquivalence'])
        plt.title("Online barista: " + str(i) + ", physical barista: " + str(5-i))
        plt.show()
        




```
